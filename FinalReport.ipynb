```python
list_of_0_99    = os.listdir('./0~99')
list_of_100_199 = os.listdir('./100~199')
list_of_200_299 = os.listdir('./200~299')
list_of_300_399 = os.listdir('./300~399')
list_of_400_499 = os.listdir('./400~499')
```

```python
def txt_to_CitationSeries_with_P(i):
    print(i)
    df1 = pd.read_csv(i, sep='\t')
    df1.columns=['a']
    citation = df1[df1['a'].str.contains('UT WOS:')].reset_index(drop=True).apply(lambda x: str(x['a'])[7:],axis=1)
    return citation
```

```python


def txt_to_CitationSeries(i):
    print(i)
    df1 = pd.read_csv(i, sep='\t')
    citation = df1[['UT']].apply(lambda x: str(x['UT'])[4:],axis=1)
    return citation
```

```python
citations = pd.DataFrame(None)

for i in list_of_0_99:
    try:
        citation = txt_to_CitationSeries('./0~99/'+i)
    except:
        citation = txt_to_CitationSeries_with_P('./0~99/'+i)
    else:
        a=1
    citation  = pd.DataFrame(citation.rename(str(i)[:-4])).T
    citations = pd.concat([citations,citation])

for i in list_of_100_199:
    try:
        citation = txt_to_CitationSeries('./100~199/'+i)
    except:
        citation = txt_to_CitationSeries_with_P('./100~199/'+i)
    else:
        a=1
    citation  = pd.DataFrame(citation.rename(str(i)[:-4])).T
    citations = pd.concat([citations,citation])
for i in list_of_200_299:
    try:
        citation = txt_to_CitationSeries('./200~299/'+i)
    except:
        citation = txt_to_CitationSeries_with_P('./200~299/'+i)
    else:
        a=1
    citation  = pd.DataFrame(citation.rename(str(i)[:-4])).T
    citations = pd.concat([citations,citation])

for i in list_of_300_399:
    try:
        citation = txt_to_CitationSeries('./300~399/'+i)
    except:
        citation = txt_to_CitationSeries_with_P('./300~399/'+i)
    else:
        a=1
    citation  = pd.DataFrame(citation.rename(str(i)[:-4])).T
    citations = pd.concat([citations,citation])

for i in list_of_400_499:
    try:
        citation = txt_to_CitationSeries('./400~499/'+i)
    except:
        citation = txt_to_CitationSeries_with_P('./400~499/'+i)
    else:
        a=1
    citation  = pd.DataFrame(citation.rename(str(i)[:-4])).T
    citations = pd.concat([citations,citation])

print(citations)
```

```python
citations.to_csv('citations.csv',encoding='UTF-8')
```

```python
Graph = nx.DiGraph()
Graph.add_nodes_from(list(citations.index))
```

```python
for i in list(citations.index):
    x = list(citations.ix[str(i),:])
    x = [j for j in x if str(j) != 'nan']
    edgelist = [(str(i),k) for k in x]
    Graph.add_edges_from(edgelist)
```

```python
plt.figure(figsize=(900,900))
pos = nx.spring_layout(Graph)
nx.draw(Graph, pos, fontsize=0, nodesize=50, with_labels=True)
plt.show()
```
